```python
# 페이지에 요청하기 위한 용도
import requests as req
# html 데이터를 만들기 위한 용도
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
url = 'https://finance.naver.com/marketindex/exchangeList.naver'
res = req.get(url)
soup = bs(res.text,'lxml')
title = soup.select('td.tit > a')
sale = soup.select('td.sale')

print(len(title))
print(len(sale))
```

```python
title_list = []
sale_list = []
for i in range(len(title)):
    title_list.append(title[i].text.strip())
    sale_list.append(sale[i].text.strip())
print(title_list)
print(sale_list)
```

```python
dic = {'통화명': title_list, '매매기준율':sale_list}
 pd.DataFrame(dic)
```
