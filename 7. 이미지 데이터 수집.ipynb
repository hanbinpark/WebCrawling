```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
import time
from selenium.webdriver.common.keys import Keys

# 파일시스템을 위한 라이브러리(삭제, 생성)
import os
# 이미지의 경로를 실제 파일로 저장
from urllib.request import urlretrieve
```

```python
# 폴더 생성
# 바탕화면에 이미지라는 폴더가 없다면 바탕화면에 이미지라는 폴더를 만들어라
if not os.path.isdir('C:/Users/gjaischooll/Desktop/이미지'):
    os.mkdir('C:/Users/gjaischooll/Desktop/이미지')
```

```python
driver = wb.Chrome()
driver.get('https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%ED%91%B8%EB%B0%94%EC%98%A4')
```

```python
# 스크롤 하는 코드
for i in range(10):
    body = driver.find_element(By.TAG_NAME,"body")
    body.send_keys(Keys.END)
```

```python
# 이미지 태그들을 수집
img = driver.find_elements(By.CSS_SELECTOR,'._image._listImage')

# 태그 안에 있는 속성을 가지고 올 때는 get_attribute('속성')
img[0].get_attribute('src')
```

```python
# src만 담아줄 리스트를 제작
src = []
for i in range(len(img)):
    src.append(img[i].get_attribute('src'))
src
```

```python
# 이미지의 경로를 실제 파일로 저장
for i in range(len(src)):
    urlretrieve(src[i],'C:/Users/gjaischooll/Desktop/이미지/'+ str(i) +'.jpg')
```
